```python
import pandas as pd
filepath = 'datafiles/AppleStore.csv'
mobilestore= pd.read_csv(filepath)
mobilestore
```

```python
#1 List of all unique prime _Genres(categories) in the data set
def primeUnique():
    ustore=mobilestore["prime_genre"].unique()
    for i in ustore:
        print(i)
    
primeUnique()
```

```python
#2 Category with highest number of app
def highestAppCount():
    high={}
    for i in mobilestore["prime_genre"]:
        if i not in high:
            high[i]=1
        else:
            high[i]+=1
    print(max(high,key=high.get),high[max(high,key=high.get)])
highestAppCount()        
```

```python
#3 Category with lowest number of app
def lowestAppCount():
    low={}
    for i in mobilestore["prime_genre"]:
        if i not in low:
            low[i]=1
        else:
            low[i]+=1
    print(min(low,key=low.get),low[min(low,key=low.get)])

lowestAppCount()  
```

```python
#4 Category using highest user rating
def categoryHigh():
    maxrate={}
    highrate=mobilestore.loc[(mobilestore['user_rating']==max(mobilestore['user_rating']))]
    for i in highrate["prime_genre"]:
        if i not in maxrate:
            maxrate[i]=1
        else:
            maxrate[i]+=1
    highestrating=max(maxrate,key=maxrate.get)
    highestcategory=maxrate[max(maxrate,key=maxrate.get)]
    print(highestrating,highestcategory)
            
categoryHigh()            
```

```python
#5 App with highest downloads
def highdownload():
    download={}
    for i in range(len(mobilestore['track_name'])):
        download[mobilestore['track_name'][i]]=mobilestore['rating_count_tot'][i]
    highdownloads=max(download,key=download.get)
    highdownloadnumber=download[max(download,key=download.get)]
    print(highdownloads,highdownloadnumber)
highdownload()   
```

```python
#7 Average user rating for free apps
def averagerating():
    freeapp=mobilestore.loc[(mobilestore['price']==0)]
    ratingavg=sum(freeapp['user_rating'])
    print(ratingavg/len(freeapp['user_rating']))

averagerating()
```

```python
#8 Average user rating for paid apps
def paidaveragerating():
    paidapp=mobilestore.loc[(mobilestore['price']!=0)]
    paidratingavg=sum(paidapp['user_rating'])
    print(paidratingavg/len(paidapp['user_rating']))

paidaveragerating()
```

```python
#10  Most frequent price point >0
def pricegreaterthanzero():
    temp=mobilestore.loc[(mobilestore["price"]>0)]
    mostfrequent=max(temp["price"].value_counts())
    print("Most frequent price point >0:",mostfrequent)
pricegreaterthanzero()
```

```python
#11  Compare average user rating for paid vs free gaming apps
def freegamingapp():
    freegaming = mobilestore.loc[(mobilestore["price"]==0) & (mobilestore["prime_genre"]=="Games")]
    paidgaming = mobilestore.loc[(mobilestore["price"]>0) & (mobilestore["prime_genre"]=="Games")]
    freerating = sum(freegaming["user_rating"])/len(freegaming)
    paidrating = sum(paidgaming["user_rating"])/len(paidgaming)
    print("free:",freerating)
    print("paid",paidrating)
    #print(abs(freerating-paidrating))
freegamingapp()    

```
