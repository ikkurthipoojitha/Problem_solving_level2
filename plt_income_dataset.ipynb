```python
# Plot Income of Arkansas from 2005 to 2014
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv('datafiles/income.csv')
years=df.columns[2:]
plt.xlabel("years")
plt.ylabel("income")
plt.title("Income of Arkansas")
plt.grid()
incomeArkansas = df.values[3,2:]
plt.plot(years,incomeArkansas)
plt.show()
```

```python
# Income of all states from 2005 to 2013
df
incomeall = []
for i in df.columns[2:]:
    incomeall.append(df[i].mean())
plt.plot(years,incomeall)
plt.grid()
'''plt.xlabel("years")
plt.plot("states income")
plt.title("income of all states with years")'''
plt.savefig("images/incomeallstates")
plt.show()
```
