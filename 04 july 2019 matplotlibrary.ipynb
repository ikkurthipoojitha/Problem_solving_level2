```python
# Linear graph
import matplotlib.pyplot as plt
plt.title("Basic Linear Graph")
plt.plot([1,2,3],[1,2,3])
plt.show()
```

```python
# Graph to plot distance and time

distance = [2,4,6,8,10]
time=[1,2,3,4,5]
plt.xlabel('Distance')
plt.ylabel('Time')
d= 2*time
plt.plot(time,distance,'ro')
plt.grid()

plt.show()
```

```python
# v= a t
t=list(range(1,11))
a=list(range(1,len(t)+1))
v=[a[i] * i for i in range(len(t))]
plt.plot(t,v)
plt.grid()
plt.show()
   
```

```python
# Numbers versus squares
n=list(range(11))
sq = [i ** 2 for i in n]
plt.plot(n,sq)
plt.show()
```

```python
# Voltage - Time (sinu-soidal)
import numpy as np
t=[i/10 for i in range(1,21)]
v=[1 + np.sin(2*np.pi*i) for i in t]
plt.plot(t,v)
plt.xlabel("time")
plt.ylabel("voltage")
plt.title("Voltage versus Time Graph")
plt.grid()
plt.savefig('images/Voltage-Time.png') # to save the graph as image
plt.show()
```
